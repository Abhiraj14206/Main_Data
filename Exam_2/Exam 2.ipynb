# Outliers

### Z Socre Method for bmi column

```python
sns.distplot(df["bmi"])
```

Both side tail that means both side outlier present.

```python
sns.boxplot(df["bmi"])
```

Positive and Negative both outlier.

```python
from scipy.stats import zscore
df["bmi_zscore"] = zscore(df["bmi"])
```

```python
df
```

```python
sns.boxplot(df["bmi_zscore"])
```

```python
outliers = df[np.abs(df["bmi_zscore"]) > 3]
outliers[["bmi", "bmi_zscore"]]
```

```python
outlier_ind = df[np.abs(df["bmi_zscore"]) > 3].index
fin_df = df.drop(outlier_ind)
```

```python
sns.boxplot(fin_df["bmi_zscore"])
```

```python
fin_df
```

```python
# Outlier Remove Using the Z-score Method
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df["bmi"],kde=True)
plt.title("Before")

plt.subplot(2,2,2)
sns.histplot(fin_df["bmi"],kde=True)
plt.title("After")

```

The Z-score method successfully removed extreme BMI outliers, reducing positive skewness and making the data more normally distributed, which improves the reliability of statistical analysis and machine learning models.

### IQR Method for Cholesterol column

```python
df.shape
```

```python
sns.boxplot(df["cholesterol"])
```

Both side Outlier.

```python
sns.histplot(df["cholesterol"],kde=True)
```

positive skew.

```python
q1 = df["cholesterol"].quantile(0.25)
q3 = df["cholesterol"].quantile(0.75)
print("Q1:",q1)
print("Q3:",q3)
```

```python
iqr = q3 -q1
print("IQR:",iqr)
```

```python
min = q1 - 1.5 * iqr
max = q3 + 1.5 * iqr
min,max
```

```python
outliers_Cho = df[(df["cholesterol"] < min) | (df["cholesterol"]> max)].index
outliers_Cho
```

```python
fin_df = df.drop(outliers_Cho)
```

```python
fin_df
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df["cholesterol"], kde=True)
plt.title("Before")

plt.subplot(2,2,2)
sns.histplot(fin_df["cholesterol"], kde=True)
plt.title("After")
```

The Cholesterol distribution is right-skewed and after removing the Few Outliers its more Symmetric.

### Percentile Method for Glucose Column

```python
df
```

```python
df.shape
```

```python
sns.boxplot(df["glucose"])
```

```python
low = df["glucose"].quantile(0.01)
high = df["glucose"].quantile(0.99)
low,high
```

```python
age_glucose = ["age","glucose"]

for col in age_glucose:
    lower = df["glucose"].quantile(0.01)
    higher = df["glucose"].quantile(0.99)
    fin_df = df[(df["glucose"]>=lower) & (df["glucose"]<=higher)]
fin_df
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
sns.histplot(df[age_glucose], kde=True)
plt.title("Before")

plt.subplot(2,2,2)
sns.histplot(fin_df[age_glucose],kde=True)
plt.title("After")
```

### Winsorization

```python
wind_df = df.copy()
numeric = ["age","bmi","cholesterol","glucose"]
for col in numeric:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    wind_df[col] = np.clip(wind_df[col], lower, upper)
wind_df
```

```python
# Boxplot comparision Before vs After
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numeric])
plt.title("Before Cleaning")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(15, 6))
sns.boxplot(data=wind_df[numeric])
plt.title("After Cleaning (Winsorization)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histplot Before vs After
plt.figure(figsize=(15, 6))
sns.histplot(data=df[numeric],kde=True)
plt.title("Before Cleaning")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(15, 6))
sns.histplot(data=wind_df[numeric],kde=True)
plt.title("After Cleaning (Winsorization)")
plt.xticks(rotation=45)
plt.show()
```
