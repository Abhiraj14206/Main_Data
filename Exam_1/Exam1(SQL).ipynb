```python
# API URL
url = "https://api.data.gov.in/resource/9a362ec2-2cfc-4e08-8c74-7926b2159a69?api-key=579b464db66ec23bdd0000016ed11d232c0b4c814d6cd6d9924a230a&format=json"

# Get the data
response = requests.get(url)
data = response.json()

# Extract only id and name from the "field" list
fields = data.get("field", [])
id_name_list = [{"id": item["id"], "name": item["name"]} for item in fields]

# Save to CSV file
with open("id_name.csv", "w", newline='', encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=["id", "name"])
    writer.writeheader()  # write column headers
    writer.writerows(id_name_list)  # write all rows

print("CSV file 'id_name.csv' created with id and name!")
```
