```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

df["Age"] = df["Age"].astype("float64")  # pandas nullable int

```

```python
Age_mean = df["Age"].mean()
print("Age Mean:",Age_mean)
df["Age"] = df["Age"].fillna(Age_mean)
df["Age"]
```

```python
df["Age"].isnull().sum()
```

```python
Gender_mode = df["Gender"].mode()[0]
Gender_mode
df["Gender"] = df["Gender"].fillna(Gender_mode)
df["Gender"]
```

```python
df.isnull().sum()
```

```python
df["Monthly_Income"] = df["Monthly_Income"].fillna(df["Monthly_Income"].mean())
df["Total_Purchases"] = df["Total_Purchases"].fillna(df["Total_Purchases"].mean())
df["Avg_Monthly_Spend"] = df["Avg_Monthly_Spend"].fillna(df["Avg_Monthly_Spend"].mean())
df["Tenure_Months"] = df["Tenure_Months"].fillna(df["Tenure_Months"].mean())
```

```python
df.isnull().sum()
```

```python
df["Churn"].unique()
```

```python
df["Churn"].value_counts()
```

```python
Chrun_mode = df["Churn"].mode()[0]
```

```python
df["Churn"] = df["Churn"].fillna(Chrun_mode)
```

```python
df['Monthly_Income'].describe()
```

# Univariate 

```python
plt.hist(df["Age"],bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows that customers are distributed across different age groups, with most customers concentrated in the middle-age range.

```python
sns.histplot(df["Monthly_Income"],bins= 20, kde=True)
```

It's Not the Normal Distribution.

```python
# df["Total_Purchases"].plot(kind="box")
sns.boxplot(x="Total_Purchases", data=df)
```

No Outlier here in the Total_Purchases.

# Bivariate

```python
sns.boxplot(x=df["Gender"],y=df["Total_Purchases"])
plt.title("Total Purchases vs Gender")
```

The boxplot shows that purchase behavior differs slightly High Female genders.

```python
plt.scatter(df["Age"],df["Total_Purchases"])
plt.xlabel("Age")
plt.ylabel("Total Purchases")
plt.title("Age vs Total Purchases")
```

This code filters your data to include only customers aged 30â€“50 and then plots their Age against Total Purchases in a clear scatter plot with labels and a title.

```python
plt.bar(df["Churn"],df["Monthly_Income"])
plt.title("Churn vs Income")
```

The bar plot shows not differences in income levels between churned and non-churned customers.

# Multivariate

```python
# Numeric column Heatmap
corr = df.corr(numeric_only= True)
sns.heatmap(corr,annot= True)
```

The heatmap shows relationships among numerical variables.

```python
high_income = df[df["Monthly_Income"] > 50000]

sns.pairplot(
    high_income[
        ['Age','Monthly_Income','Total_Purchases',
         'Avg_Monthly_Spend','Tenure_Months']
    ]
)

plt.show()
```

```python
sns.pairplot(df[['Age','Monthly_Income','Total_Purchases','Avg_Monthly_Spend','Tenure_Months']])
plt.show()
```

The pair plot confirms patterns and relationships between the Age, Monthly_Income, Total_purchases, Avg_monthly_spend, Tensure_months.
